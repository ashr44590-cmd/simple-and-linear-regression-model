```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Create dataset
data = {
    'Hours':[1,2,3,4,5,6,7,8],
    'Marks':[40,45,50,60,65,70,75,85]
}

df = pd.DataFrame(data)

# Independent and dependent variables
X = df[['Hours']]
y = df['Marks']

# Create model
model = LinearRegression()

# Train model
model.fit(X,y)

# Prediction
predicted = model.predict(X)

# Plot graph
plt.scatter(X,y)
plt.plot(X,predicted)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Simple Linear Regression")
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Dataset
data = {
    'Hours':[2,3,4,5,6,7],
    'Attendance':[70,75,80,85,90,92],
    'Sleep':[6,7,6,7,8,7],
    'Marks':[50,55,60,65,70,75]
}

df = pd.DataFrame(data)

# Independent variables
X = df[['Hours','Attendance','Sleep']]

# Dependent variable
y = df['Marks']

# Model
model = LinearRegression()

# Train
model.fit(X,y)

# Prediction
prediction = model.predict([[5,85,7]])

print("Predicted Marks:", prediction)
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

# Data
hours = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)
marks = np.array([40,45,50,60,65,70,75,85])

# Model
model = LinearRegression()
model.fit(hours, marks)

# Prediction line
pred = model.predict(hours)

# Graph
plt.scatter(hours, marks)
plt.plot(hours, pred)

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression Graph")

plt.show()
```
